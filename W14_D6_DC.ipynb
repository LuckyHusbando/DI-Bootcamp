The `openpyxl` library is now installed. Next, I will define the student grade data and use `openpyxl` to create and populate the Excel workbook.

```python
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill

# Data structure provided by the user
data = {
    "Joe": {
        "math": 65,
        "science": 78,
        "english": 98,
        "gym": 89
    },
    "Bill": {
        "math": 55,
        "science": 72,
        "english": 87,
        "gym": 95
    },
    "Tim": {
        "math": 100,
        "science": 45,
        "english": 75,
        "gym": 92
    },
    "Sally": {
        "math": 30,
        "science": 25,
        "english": 45,
        "gym": 100
    },
    "Jane": {
        "math": 100,
        "science": 100,
        "english": 100,
        "gym": 60
    }
}

# Create a new workbook and select the active sheet
wb = Workbook()
ws = wb.active
ws.title = "Grades"

# Define the header row
headers = ["Student Name", "math", "science", "english", "gym"]
ws.append(headers)

# Apply bold and colored formatting to the header row
header_font = Font(bold=True)
header_fill = PatternFill(start_color="ADD8E6", end_color="ADD8E6", fill_type="solid") # Light blue color

for col_idx in range(1, len(headers) + 1):
    cell = ws.cell(row=1, column=col_idx)
    cell.font = header_font
    cell.fill = header_fill

# Populate the sheet with student data
for student_name, grades in data.items():
    row_data = [student_name, grades["math"], grades["science"], grades["english"], grades["gym"]]
    ws.append(row_data)

# Add a row for average grades. The formulas will start from the second column (B).
# The data starts from row 2 (after the header) and goes up to the current last row - 1 (before the average row).
num_students = len(data)

# Cell 'B1' is the 'math' header, so data starts from 'B2'
# The average row will be `num_students + 2`
for col_idx, subject in enumerate(headers[1:], start=2): # Start from column B (index 2)
    # The range for the average formula will be from row 2 to num_students + 1 (the last student's data row)
    start_cell = f"{chr(64 + col_idx)}2"
    end_cell = f"{chr(64 + col_idx)}{num_students + 1}"
    formula = f"=AVERAGE({start_cell}:{end_cell})"
    ws.cell(row=num_students + 2, column=col_idx, value=formula)

# Add a label for the average row
ws.cell(row=num_students + 2, column=1, value="Average")

# Save the workbook
excel_file_path = "student_grades.xlsx"
wb.save(excel_file_path)

print(f"Excel workbook '{excel_file_path}' created successfully.")
```
